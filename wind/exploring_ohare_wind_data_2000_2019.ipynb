## Cleaning the daily wind data for Ohare airport for the years 2000-2019.

```python
import pandas as pd
```

```python
df = pd.read_csv('../data/Wind_data/ohare_wind_2000_2019_unclean.csv')
print(df.shape)
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['FMTM', 'PGTM', 'STATION', 'NAME'])
```

```python
ohare_wind = df.drop(columns=['FMTM', 'PGTM', 'STATION', 'NAME'])
```

```python
ohare_wind['DATE'] = pd.to_datetime(ohare_wind['DATE'])
```

```python
ohare_wind['DATE'].nunique()
```

```python
ohare_wind.set_index('DATE', inplace=True)
```
