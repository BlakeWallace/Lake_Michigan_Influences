```python
data_lst = soup.text.split()
```

```python
# extracting data (full data set)

keys = []
values = []
for i in range(len(data_lst)):
    if i % 2 == 0:
        keys.append(int(data_lst[i]))
    else:
        values.append(float(data_lst[i]))
```

```python
keys
```

```python
values
```

```python
data_dict = {}
for i in range(len(values)):
    data_dict[keys[i]] = values[i]
data_dict
```

```python
pd.DataFrame(data_dict, columns=['daily_avg'])
```

```python
data_dict = {
    '1992-2014_daily_avg': values
}
```
