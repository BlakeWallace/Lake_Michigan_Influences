### Datasets included:
 - Fsti2, Foster Beach buoy data
  
 - O'Hare weather data
  
 - Lake Michigan Daily average temperatures

```python
# buoy FSTI2
buoy = pd.read_csv('data/buoy_FSTI2/jan012015_dec312018.csv')
print(buoy.shape)
buoy.head()
```

```python
buoy.rename(mapper={'WDIR': 'buoy_wdir',
                    'WSPD': 'buoy_wspd',
                    'ATMP': 'buoy_atmp'}, 
            axis=1, 
            inplace=True)

buoy.head()
```

```python
# ohare weather data (Jan 01, 2015 to Dec 31, 2018)

ohare = pd.read_csv('data/ohare/jan012015_dec312018.csv')
print(ohare.shape)
ohare.head()
```

```python
ohare.drop(columns=['WSF2', 'WDF2'], inplace=True)
ohare.head()
```

```python
ohare.rename(mapper={'AWND': 'ohare_wspd',
                     'PRCP': 'ohare_prcp',
                     'SNOW': 'ohare_snfall',
                     'SNWD': 'ohare_sndpth',
                     'TAVG': 'ohare_atmp',
                     'TMAX': 'ohare_Mtmp',
                     'TMIN': 'ohare_mtmp'},
             axis=1,
             inplace=True)

ohare.head()
```

```python
# lake michigan daily average temps
lake_mic = pd.read_csv('data/lake_michigan/jan012015_dec312018.csv')
print(lake_mic.shape)
lake_mic.head()
```

```python
data = pd.concat([lake_mic, buoy.drop(columns='DATE'), ohare.drop(columns='DATE')], axis=1)
print(data.shape)
data.head()
```

```python
# saving the data
data.to_csv('data/merged/jan012015_dec312018.csv', index=False)
```
