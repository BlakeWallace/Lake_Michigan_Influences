## In this notebook we bring together the data from all of the datasets.

### Datasets included:
 - 2015-2018 buoy data from
  - Fsti2, Foster Beach
  - Jaki2, 63rd St
  - Chii2, Submerged off the coast
  
 - Ohare Wind data, 2000-2018
 - Ohare Temp data, 2000-2018
 - Ohare Precipitation data, 2000-2018
  
 - Lake Michigan Daily average temperatures, 2000-2018

### We begin by combining all the datasets for the years 2015-2018.

```python
import pandas as pd
import time
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
# buoy FSTI2
fsti = pd.read_csv('data/buoy_FSTI2/fsti2_buoy_data_2015_2018_averages.csv')
print(fsti.shape)
fsti.head()
```

```python
# buoy JAKI2
jaki = pd.read_csv('data/buoy_JAKI2/jaki2_buoy_data_2015_2018_averages.csv')
jaki = jaki.drop(columns=['GST', 'DEWP'])
print(jaki.shape)
jaki.head()
```

```python
# buoy CHII2
chii_with_GST = pd.read_csv('data/buoy_CHII2/chii2_buoy_data_2015_2018_averages.csv')
chii = chii_with_GST.drop(columns='GST')
print(chii.shape)
chii.head()
```

```python
# Ohare wind
ohare_wind_2000_2019 = pd.read_csv('data/Wind_data/ohare_wind_2000_2019.csv')
print(ohare_wind_2000_2019.shape)
ohare_wind_2000_2019.head()
```

```python
# Ohare temp
ohare_temp_2000_2019 = pd.read_csv('data/temp_data/ohare_temp_2000_2019.csv')
print(ohare_temp_2000_2019.shape)
ohare_temp_2000_2019.head()
```

```python
# lake michigan daily average temps
lake_mic = pd.read_csv('data/temp_data/lake_michigan_water_temp_1995_2019.csv')
print(lake_mic.shape)
lake_mic.head()
```
