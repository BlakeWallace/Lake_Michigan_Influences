```python
# column titles

# magic code to get the column titles
# soup.find_all('td', {'class':'coltitle', 'width': '64'})

col_titles = []
for row in soup.find_all('td', {'class':'coltitle', 'width': '64'}):
    col_titles.append(row.text)
col_titles.append('Month')



# the beginning of the code for getting the data from this website, 
# https://www.weather.gov/lot/January_Normals_Chicago.

months = ['January', 'February', 'March', 'April', 'May', 'June',
          'July', 'August', 'September', 'October', 'November', 'December']
title = []

columns = [col for col in col_titles]

day_of_month = []
Month = []

for i in ['January', 'February']: #months:
    url = f'https://www.weather.gov/lot/{i}_Normals_Chicago'
    
    res = requests.get(url)
    
    soup = BeautifulSoup(res.content, 'lxml')
    
    # First column, 'Day of Month', January, February

    
    for data in soup.find_all('td', {'align':'center', 'height':'20'}):
        if len(data.text) <= 2:
            day_of_month.append(data.text)
            Month.append(i)

    print(soup.find('h1').text)
```
